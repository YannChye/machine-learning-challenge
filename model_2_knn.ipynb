```python
# import dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
# this jupyter notebook will use the k-nearest neigbour classifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.feature_selection import mutual_info_classif
```

# Read the CSV and Perform Basic Data Cleaning

```python
# read in data
df = pd.read_csv("./data/exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head()
```

# Select your features (columns)

```python
# Set features (removing features that are anticorrelated such as 'koi_period_err1' & 'koi_period_err2')
# features removed where based on examination of correlation matrix during running of model_1_rf
X1 = df[['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec','koi_period','koi_period_err1','koi_time0bk','koi_time0bk_err1','koi_impact','koi_impact_err1','koi_impact_err2','koi_duration','koi_duration_err1','koi_depth','koi_depth_err1','koi_prad','koi_prad_err1','koi_prad_err2','koi_teq','koi_insol','koi_insol_err1','koi_insol_err2','koi_model_snr','koi_tce_plnt_num','koi_steff','koi_steff_err1','koi_steff_err2','koi_slogg','koi_slogg_err1','koi_slogg_err2','koi_srad','koi_srad_err1','koi_srad_err2','ra','dec','koi_kepmag']]
feature1_name = ['koi_fpflag_nt','koi_fpflag_ss','koi_fpflag_co','koi_fpflag_ec','koi_period','koi_period_err1','koi_time0bk','koi_time0bk_err1','koi_impact','koi_impact_err1','koi_impact_err2','koi_duration','koi_duration_err1','koi_depth','koi_depth_err1','koi_prad','koi_prad_err1','koi_prad_err2','koi_teq','koi_insol','koi_insol_err1','koi_insol_err2','koi_model_snr','koi_tce_plnt_num','koi_steff','koi_steff_err1','koi_steff_err2','koi_slogg','koi_slogg_err1','koi_slogg_err2','koi_srad','koi_srad_err1','koi_srad_err2','ra','dec','koi_kepmag']
```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
# dummy-encode categorical outcome
label_encoder = LabelEncoder()

label_encoder.fit(df['koi_disposition'])
y=label_encoder.transform(df['koi_disposition'])
```

```python
# create train/test split
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, random_state=1)
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
# Scale the data features
X1_scaler = MinMaxScaler().fit(X1_train)
X1_train_scaled = X1_scaler.transform(X1_train)
X1_test_scaled = X1_scaler.transform(X1_test)
```

```python
# using mutual info classifier to determine the dependencies between features and outcome, 
# using n-neighbors=5 as that provided the best test score,
# to select a subset of importance features
mutual_info_classif(X1_train_scaled,y_train,random_state=5)
```

```python
# new feature set with a subset of features that share mutual info between feature and target of >0.1
subset=[0,1,2,4,5,7,12,13,15,16,17,22,25,26]
X2_train_scaled=X1_train_scaled[:,subset]
X2_test_scaled=X1_test_scaled[:,subset]
feature2_name=[feature1_name[i] for i in subset]
print(f'features:{feature2_name}')
```

# Hyperparameter Tuning

Use `GridSearchCV` to tune the model's parameters

```python
# rerun k-nearest neighbors classifier with new subset of features
# using GridSearch for parameter tuning
from sklearn.model_selection import GridSearchCV
leaf_size=list(range(1,20))
param_grid={'n_neighbors':[3,5,7,9,11,13,15,17,19],
            'weights':['uniform','distance'],
            'leaf_size':leaf_size,
            'p':[1,2],
           'metric':['euclidean','manhattan']}
grid = GridSearchCV(KNeighborsClassifier(),param_grid,verbose=2)
```

```python
# Train the model with GridSearch
grid.fit(X2_train_scaled, y_train)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```

```python
# final model
knn_final=KNeighborsClassifier(leaf_size=1,metric='manhattan',n_neighbors=19,p=1,weights='distance')
```

```python
knn_final = knn_final.fit(X2_train_scaled, y_train)
```

```python
# final model scores
print(f"Final model Training Data Score: {knn_final.score(X2_train_scaled, y_train)}")
print(f"Final model Testing Data Score: {knn_final.score(X2_test_scaled, y_test)}")
```
